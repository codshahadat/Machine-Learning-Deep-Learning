```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/kaggle/input/banana/banana_quality.csv')
df
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.dropna(axis=0,inplace=True)
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split 
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
le=LabelEncoder()
```

```python
# Encode Quality column
df['Quality'] = le.fit_transform(df['Quality'])

# Perform data type conversions
df[['Size', 'Weight', 'Sweetness', 'Softness', 'HarvestTime', 'Ripeness', 'Acidity']] = df[
    ['Size', 'Weight', 'Sweetness', 'Softness', 'HarvestTime', 'Ripeness', 'Acidity']].astype('float64')
```

```python
df
```

```python
x = df.drop(columns=['Quality'])
y = df.Quality
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
x
```

```python
model_svm = SVC(kernel = 'rbf', random_state = 0)
model_svm.fit(X_train, y_train)
```

```python
y_pred = model_svm.predict(X_test)
```

```python
accuracy_svm = accuracy_score(y_test, y_pred)
print('Accuracy:',accuracy_svm)
```

```python
classification_rep_svm = classification_report(y_test, y_pred)
print(classification_rep_svm)
```
