```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/kaggle/input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv')
df
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.dropna(axis=0,inplace=True)
df.isna().sum()
```

```python
df.drop(columns='id',inplace=True)
```

```python
df = df.rename(columns={'ever_married': 'Married'})
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split 
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
le=LabelEncoder()
```

```python
df.age=df.age.astype('int64')
df.avg_glucose_level=df['avg_glucose_level'].astype('int64')
df.bmi=df['bmi'].astype('int64')
df["gender"]=le.fit_transform(df["gender"])
df["Married"]=le.fit_transform(df["Married"])
df["work_type"]=le.fit_transform(df["work_type"])
df["Residence_type"]=le.fit_transform(df["Residence_type"])
df["smoking_status"]=le.fit_transform(df["smoking_status"])
```

```python
df
```

```python
x = df.drop(columns=['stroke','avg_glucose_level', 'heart_disease'])
y = df.stroke
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
x
```

```python
model_svm = SVC(kernel = 'sigmoid', random_state = 0)
model_svm.fit(X_train, y_train)
```

```python
y_pred = model_svm.predict(X_test)
```

```python
accuracy_svm = accuracy_score(y_test, y_pred)
print('Accuracy:',accuracy_svm)
```

```python
classification_rep_svm = classification_report(y_test, y_pred)
print(classification_rep_svm)
```
